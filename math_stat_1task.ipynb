```python
#Т1 - Сгенерируйте выборку объема N = 100 для конкретного значения параметра θ. Для всех значений 1 <= n <= N вычислите исправленные значения оценки и постройте графики |θk' - θ| в зависимости от n.
```

```python
import matplotlib.pyplot as plt
import numpy as np
import random

θ = 100
N = 100
```

```python
θ1 = []   # = 2 * mean(xn)
θ2 = []   # = (n + 1) * min(xn)
θ3 = []   # = (n + 1)/n * max(xn)
θ4 = []   # = min(xn) + max(xn)
θ5 = []   # = xn[1] + mean(xn[1:])
for n in range(1, N + 1):
  #xn = random.sample(range(0, θ), n)    
  xn = []
  for i in range(n):
    xn.append(random.random()*θ)  #генерируем выборку из n чисел равномерно распределенных на промежутке (0, θ)
  #print(n, ": ", xn)
  θ1.append(np.fabs(2 * np.mean(xn) - θ))
  θ2.append(np.fabs((n + 1) * np.min(xn) - θ))
  θ3.append(np.fabs((n + 1)/n * np.max(xn) - θ))
  θ4.append(np.fabs(np.min(xn) + np.max(xn) - θ))
  if n > 1:
    θ5.append(np.fabs(xn[0] + np.mean(xn[1:]) - θ))
  else:
    θ5.append(np.fabs(xn[0] - θ))
  #print(θ1[n - 1], θ2[n - 1], θ3[n - 1], θ4[n - 1], θ5[n - 1])
```

```python
fig, ax = plt.subplots(figsize = (10, 6))
fig.suptitle("θ = 200")



ax.plot(range(1, N + 1), θ1, label='θ1 = 2 * mean(xn)')
ax.plot(range(1, N + 1), θ2, label='θ2 = (n + 1) * min(xn)')
ax.plot(range(1, N + 1), θ3, label='θ3 = (n + 1)/n * max(xn)')
ax.plot(range(1, N + 1), θ4, label='θ4 = min(xn) + max(xn)')
ax.plot(range(1, N + 1), θ5, label='θ5 = xn[1] + mean(xn[1:])')
handles, labels = ax.get_legend_handles_labels()
ax.legend(handles, labels)
ax.grid(True)
#ax.minorticks_on()
#ax.grid(which='minor')
ax.set_ylabel('Отклонение от θ')
ax.set_xlabel('Размер выборки n')

plt.show()
```
